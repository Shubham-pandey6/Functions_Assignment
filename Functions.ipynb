1. Difference Between a Function and a Method in Python
Function: A block of reusable code that is called independently. Defined using def outside of any class.

````python
def add(a, b):
    return a + b

result = add(2, 3)  # result is 5
````

Method: A function that is associated with an object (usually defined inside a class and called on an instance).

Example:

````python
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(2, 3)  # result is 5
print(result)
````

5. Iterators vs Iterables in Python
Iterable: An object with elements that can be looped over (has __iter__()), e.g., lists, strings.

Iterator: An object with a __next__() method, returning items from the iterable—one at a time.

Difference: Every iterator is iterable; not every iterable is an iterator.

Example:

````python
lst = [1, 2, 3]
it = iter(lst)  # 'it' is now an iterator
next(it)        # Returns 1
````

6. Generators in Python: Concept and Definition
Generators: Functions that yield values one at a time using the yield keyword, producing a sequence of results lazily.

Example:

````python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for i in countdown(3):  # Yields 3, 2, 1
    print(i)
````

7. Advantages of Generators over Regular Functions
Memory Efficiency: Generate values one at a time; don’t store entire sequences in memory.

Lazy Evaluation: Only compute values as needed.

Example: Reading large files line by line with a generator uses less memory than loading the whole file.

1. Function to Return the Sum of All Even Numbers in a Lis

````python
def sum_even_numbers(nums):
    return sum(x for x in nums if x % 2 == 0)

# Example:
lst = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(lst))  # Output: 12
````

3. Function to Return Squares of Each Number in a List

````python
def squares_of_list(nums):
    return [x**2 for x in nums]

# Example:
nums = [1, 2, 3, 4]
print(squares_of_list(nums))  # Output: [1, 4, 9, 16]
````

4. Function to Check if a Number (1–200) is Prime or Not

````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Print all primes from 1 to 200
for i in range(1, 201):
    if is_prime(i):
        print(i, end=' ')
# Output: Prints all primes between 1 and 200
````

5. Iterator Class for the Fibonacci Sequence

````python
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example:
fib = Fibonacci(7)
for num in fib:
    print(num, end=' ')  # Output: 0 1 1 2 3 5 8
````

6. Generator Function Yielding Powers of 2 Up to a Given Exponent

````python
def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2**i

# Example:
for power in powers_of_two(5):
    print(power, end=' ')  # Output: 1 2 4 8 16 32
````

7. Generator Function to Read a File Line by Line

````python
def read_file_by_line(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.rstrip('\n')

# Example usage (Suppose 'example.txt' exists):
# for line in read_file_by_line('example.txt'):
#     print(line)
````

8. Lambda Function to Sort List of Tuples by Second Element

````python
tuples_list = [(1, 3), (2, 1), (4, 2)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)  # Output: [(2, 1), (4, 2), (1, 3)]
````

9. Use map() to Convert List of Celsius to Fahrenheit

````python
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)  # Output: [32.0, 68.0, 98.6, 212.0]
````

10. Use filter() to Remove All Vowels from a String

````python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example:
text = "Hello World"
print(remove_vowels(text))  # Output: "Hll Wrld"
````
